```python
import cv2
import numpy as np
```

```python
image = cv2.imread(r"dog_image.jpg")
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
img = cv2.blur(image, ksize = (5, 5))
```

```python
cv2_imshow(img)
```

```python
sharpen_filter=np.array([[-1,-1,-1],
                 [-1,9,-1],
                [-1,-1,-1]])
```

```python
sharp_img = cv2.filter2D(img, -1 , sharpen_filter)
```

```python
sharp_img
```

```python
cv2_imshow(sharp_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
